## Beysian Ridge Regression

```python
from sklearn.linear_model import BayesianRidge
```

```python
def objective(trial):

    param = {
        'n_iter':trial.suggest_int("n_iter", 100, 900),
        'tol':trial.suggest_uniform('tol', 1e-5, 1e-2),
        'alpha_1' : trial.suggest_uniform('alpha_1',1e-9, 1e-5),
        'alpha_2': trial.suggest_uniform('alpha_2',1e-9, 1e-5),
        'lambda_1': trial.suggest_uniform('lambda_1',1e-9, 1e-5),
        'lambda_2': trial.suggest_uniform('lambda_2',1e-9, 1e-5),
        'compute_score': trial.suggest_categorical('compute_score',[True, False]),
        'fit_intercept': trial.suggest_categorical('fit_intercept',[True, False]),
        'normalize': trial.suggest_categorical('normalize',[True, False]),
        'copy_X' :trial.suggest_categorical('copy_X',[True, False]),
        'verbose':trial.suggest_categorical('verbose',[True, False]),
    }

    bys = BayesianRidge(**param)
    model_bys  = bys.fit(X_train, y_train)

    MSE = mean_squared_error(y_test, model_bys.predict(X_test))
    return MSE
```

```python
optuna_bym = optuna.create_study(direction='maximize')
optuna_bym.optimize(objective, n_trials=100)
```

```python
bym_trial = optuna_bym.best_trial
bym_trial_params = bym_trial.params
print('Best Trial: score {},\n params {}'.format(bym_trial.value, bym_trial_params))
```

```python
# Modeling fit
bym = BayesianRidge(**bym_trial_params)
bym_study = bym.fit(X_train, y_train)

# Predict the y_test
y_pred = bym_study.predict(X_test)

score = r2_score(y_test, y_pred)
mse= mean_squared_error(y_test, y_pred)
```

```python
print("Baysian Best R2 : {0:.4f}, MSE : {1:.4f}, RMSE : {2:.4f}".format(score, mse, np.sqrt(mse)))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 5))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Height")
plt.ylabel("Predicted Height")
plt.xlim([145,195])
plt.ylim([145,195])
plt.title("BAYESIAN REGRESSION")
plt.show()
```

## CatBoost

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from catboost import CatBoostRegressor, Pool, cv
```

```python
def objective(trial):

    param = {
        'iterations':trial.suggest_int("iterations", 4000, 25000),
        'od_wait':trial.suggest_int('od_wait', 500, 2300),
        'learning_rate' : trial.suggest_uniform('learning_rate',0.01, 1),
        'reg_lambda': trial.suggest_uniform('reg_lambda',1e-5,100),
        'subsample': trial.suggest_uniform('subsample',0,1),
        'random_strength': trial.suggest_uniform('random_strength',10,50),
        'depth': trial.suggest_int('depth',1, 15),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf',1,30),
        'leaf_estimation_iterations': trial.suggest_int('leaf_estimation_iterations',1,15),
        'bagging_temperature' :trial.suggest_loguniform('bagging_temperature', 0.01, 100.00),
        'colsample_bylevel':trial.suggest_float('colsample_bylevel', 0.4, 1.0),
        'random_state' : 0
    }

    gbm = CatBoostRegressor(**param)
    model_cbrm  = gbm.fit(X_train, y_train, eval_set=[(X_test, y_test)],verbose=0, early_stopping_rounds=25)

    MSE = mean_squared_error(y_test, model_cbrm.predict(X_test))
    return MSE
```

```python
optuna_cbrm = optuna.create_study(direction='maximize')
optuna_cbrm.optimize(objective, n_trials=100)
```

```python
cbrm_trial = optuna_cbrm.best_trial
cbrm_trial_params = cbrm_trial.params
print('Best Trial: score {},\n params {}'.format(cbrm_trial.value, cbrm_trial_params))
```

```python
# Modeling fit
cbrm = CatBoostRegressor(**cbrm_trial_params)
cbrm_study = cbrm.fit(X_train, y_train)

# Predict the y_test
y_pred = cbrm_study.predict(X_test)

score = r2_score(y_test, y_pred)
mse= mean_squared_error(y_test, y_pred)
```

```python
print("Best R2 : {0:.4f}, MSE : {1:.4f}, RMSE : {2:.4f}".format(score, mse, np.sqrt(mse)))
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Height")
plt.ylabel("Predicted Height")
plt.xlim([145,195])
plt.ylim([145,195])
plt.title("CatBoost REGRESSION")
plt.show()
```
